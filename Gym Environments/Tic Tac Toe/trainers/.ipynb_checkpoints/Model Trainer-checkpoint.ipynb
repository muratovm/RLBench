```python
import gym
import gym_tictactoe


env = gym.make('tictactoe-v0', dim=3)
```

```python
from agents.ModelAgent import SmartAgent
import torch.optim as optim


agents = []
agents.append(SmartAgent("Bob", "O"))
agents.append(SmartAgent("Milo", "X"))

for agent in agents:
    env.register_agent(agent)
    
    
optimizers = []
optimizers.append(optim.Adam(agents[0].parameters(), amsgrad=True, weight_decay=0.01))
optimizers.append(optim.Adam(agents[1].parameters(), amsgrad=True, weight_decay=0.01))
```

```python
import torch 
import torch.nn as nn
```

```python
#while not env.winner:
env.reset()
done = False
criterion = nn.MSELoss()
criterion2 = nn.CrossEntropyLoss()
torch.autograd.set_detect_anomaly(True)

for i in range(150):
    while not done:
        for i in range(len(agents)):
            agent = agents[i]
            optimizer = optimizers[i]
            
            choice = agent(env.board)
            
            choice_row = torch.argmax(choice[0])
            choice_col = torch.argmax(choice[1])
            pos = [choice_row, choice_col]
            
            action = [pos, agent]


            optimizer.zero_grad()

            exp_reward = agent(env.board, pos[0], pos[1])
            reward, done = env.step(action)

            reward = torch.tensor([[reward]], dtype=torch.float32)

            loss = criterion(reward, exp_reward)
            loss.backward()
            optimizer.step()
            
            agent.trajectory.append((env.board,pos,reward))
            
            
            best_move = agent.best_reward_move(env.board)
            
            if best_move:
            
                #compare move made and best move
                #print(choice[0])
                #print(best_move[1])



                optimizer.zero_grad()
                choice = agent(env.board)
                loss = criterion2(choice[0], torch.tensor([best_move[0]]))
                #print(loss)
                loss.backward()
                optimizer.step()

                #print(choice[1])
                #print(best_move[0])

                optimizer.zero_grad()
                choice = agent(env.board)
                loss = criterion2(choice[1], torch.tensor([best_move[1]]))
                loss.backward()
                optimizer.step()
                
                
            

            #print(exp_reward[0][0].item(), reward[0][0].item())

            #env.render()
            #print("="*env.dim)
            if done: break


    if env.winner == None:
        print("---------------------")
        print("No one won!")
        print("---------------------")

    else:
        print("---------------------")
        print(agent.name+" won!")
        print("---------------------")
    
    env.render()
    env.reset()
    for agent in agents:
        agent.learn_trajectory()
    done = False
```
